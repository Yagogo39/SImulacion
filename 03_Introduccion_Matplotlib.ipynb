### Representacion Grafica de los Datos

Si a la funcion de trazado de una matriz de datos, usará como coordenadas en el eje vertical, y se utilizara el indice de cada punto de datos en el array como la coordenada horizontal.

```python
plt.plot([1,2,5,7,7,8,3,1])
plt.show()
```

Tambien se puede proporcionar dos matrices: una para el eje horizontal y otra para el eje vertical.

```python
plt.plot([-3,-1,0,4,7],[1,4,6,7,8]) #Eje x, Eje y
plt.axis([-4,4,0,10]) # [xmin, xmax, ymin,  ymax]
plt.show()
```

Pueden modificarse las longitudes de los ejes para que la figura no se vea tan ajustada.

```python
plt.plot([-3,-1,0,4,7], [1,4,6,7,8])
plt.axis([-4,4,0,10]) # [xmin, xmax, ymin, ymax]
plt.show()
```

Se sigue el mismo procedimiento para pintar una funcion matematica

```python
import numpy as np
x = np.linspace(-2, 2, 500) #500 puntos entre -2 y 2
y=x**2 #y es x al cuadrado

plt.plot(x,y) #Grafica y vs x
plt.show()
```

Tambien puede modificarse el estilo de la grafica para que contenga mas informacion

```python
plt.plot(x,y)
plt.title("Square function") #Titulo de la grafica
plt.xlabel("Matematico") #Etiqueta del eje x
plt.ylabel("y=x**2") #Etiqueta del eje y
plt.grid(True) #Para que se vea una cuadricula
plt.show()
```

Pueden superponerse graficas y cambiar el estilo de las lineas

```python
x=np.linspace(-2,2,500) #500 puntos entre -2 y 2
y=x**2 #y es x al cuadrado
y2=x+1

plt.plot(x,y, 'b--', x, y2, 'g') #Primera grafica azul punteada, segunda grafica verde solida
plt.show()
```

```python
#Separando en diferentes lineas las funciones
x=np.linspace(-2,2,500) #500 puntos entre -2 y 2
y=x**2 #y es x al cuadrado
y2=x+1

plt.title("Square function") #Titulo de la grafica
plt.plot(x,y, 'b--') #Primera grafica azul punteada
plt.plot(x,y2, 'g') #Segunda grafica verde solida
plt.show()
```

Para poder diferenciar entre las funciones siempre es recomendable añadir una leyenda

```python
x=np.linspace(-2,2,500)
y=x**2 #y es x al cuadrado
y2=x+1

plt.title("Square function") #Titulo de la grafica
plt.plot(x,y, 'b--', label='y=x^2') #Primera grafica azul punteada
plt.plot(x,y2, 'g', label='y=x+1') #Segunda grafica verde solida
plt.legend(loc='best')  #Muestra la leyenda
plt.show()
```

Tambien puede crearse dos graficas que no se suporpongan. Estas graficas se organizan en un grid y se denominan subplots.

```python
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1) #1 fila, 2 columnas, primera grafica
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) #1 fila, 2 columnas, segunda grafica
plt.plot(x, y2, 'g')

plt.show()
```

#### Scatter Plots

```python
from numpy.random import rand

x, y = rand(2, 100) #Genera un arreglo de 2 filas y 100 columnas con numeros aleatorios entre 0 y 1
plt.scatter(x, y) #Grafica de dispersion
plt.show()
```

Dos graficas de dispersion en un mismo grid

```python
x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c='red')
plt.scatter(x2, y2, c='blue')
plt.show()
```

#### Histogramas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(2,1,1) #2 filas, 1 columna, primera grafica
plt.hist(data, bins=10, rwidth=0.8) #bins es el numero de barras, rwidth es el ancho relativo de las barras

plt.subplot(2,1,2) #2 filas, 1 columna, segunda grafica
plt.hist(data, bins=[1,1.5,2,2.5,3], rwidth=0.95)

plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```
